# **Fertilitätsrate nach Alter und Herkunft der Mutter**

## Description


Fertilitätsraten nach Jahr, Alter und Herkunft der Mutter.


**Fertilitätsrate**

Die altersspezifische Fertilitätsrate beschreibt das Verhältnis der Anzahl Geburten von Müttern eines Alters zu der Gesamtzahl Frauen der Bevölkerung im entsprechenden Alter. Sie wird oft in Promille angegeben und entspricht dann der durchschnittlichen Anzahl Geburten pro 1000 Frauen eines bestimmten Alters.
 

[]()

**Lebendgeborene**

Geborene gelten als lebend geboren, wenn sie nach völligem Austritt aus dem Mutterleib spontan atmen und Herzschläge aufweisen; andernfalls gelten sie als tot geboren, sofern ihr Geburtsgewicht mindestens 500 Gramm oder ihr Alter mindestens 22 vollendete Schwangerschaftswochen beträgt. Hier werden bei den Geburten nur Lebendgeborene ausgewiesen.


[]()

**Fehlende Werte**

Es sind nicht in allen Jahren Daten verfügbar. Vor allem die Anzahl 15- bis 49-jähriger Frauen ist nicht für alle Jahre vorhanden.


[]()

**Altersklassen**

Hier werden Zeitreihen zu Fertilitätsraten erstellt. Als Datenquellen wurden vor 1993 unter anderem Statistische Jahrbücher verwendet. Dort gibt es bei den Geburtenzahlen teilweise nur eine und nicht zwei Altersgrenzen. Beispiel: Geburten 45-Jährigen und Älteren. Für die Fertilitätsrate macht es wenig Sinn, die Geburten der 45-Jährigen und Älteren durch die Anzahl aller 45- bis 100-jähriger oder noch älterer Frauen zu teilen. Darum werden die Geburten der 45-Jährigen und Älteren durch die Anzahl 45- bis 49-jähriger Frauen geteilt. Bei der Altersklasse 45 bis 49, kann es jedoch sein, dass einige Kinder durch 50-jährige und ältere Frauen geboren wurden. Analog ist es bei den 15- bis 19-Jährigen: Es kann sein, dass einige Kinder durch 14-Jährige und Jüngere geboren wurden.

## Imports and helper functions

```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# helper function for reading datasets with proper separator
def get_dataset(url):
    """
    Return pandas df if url is parquet or csv file. Return None if not.
    """
    extension = url.rsplit(".",1)[-1]
    if extension == 'parquet':
        data = pd.read_parquet(url)
    elif extension == 'csv':
        data = pd.read_csv(url, sep=",", on_bad_lines='warn', encoding_errors='ignore', low_memory=False)
        # if dataframe only has one column or less the data is not comma separated, use ";" instead
        if data.shape[1] <= 1:
            data = pd.read_csv(url, sep=';', on_bad_lines='warn', encoding_errors='ignore', low_memory=False)
            if data.shape[1] <= 1:
                print("The data wasn't imported properly. Very likely the correct separator couldn't be found.\nPlease check the dataset manually and adjust the code.")
    else:
        print("Cannot load data! Please provide an url with csv or parquet extension.")
        data = None
    return data
```
