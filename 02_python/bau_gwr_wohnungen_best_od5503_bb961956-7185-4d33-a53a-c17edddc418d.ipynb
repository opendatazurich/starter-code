## Description

Der vorliegende Datensatz enthält täglich aktualisierte Informationen zu allen Wohnungen in der Stadt Zürich, gemäss dem GWR-Datenmodell.

Bitte beachten Sie für das korrekte Verständnis zu den Daten die unten aufgeführten weiteren Erläuterungen in der Rubrik **«Bemerkungen»**.

**GWR-Datenmodell:**

Die hier zur Verfügung gestellten Daten stammen aus dem **Gebäude- und Wohnungsregister der Stadt Züirich** (GWZ), entsprechen strukturell jedoch dem vom **Eidgenössischen Gebäude- und Wohnungsregister** (GWR) definierten Datenmodell. 

Im GWR werden nebst den schweizweit eindeutigen Gebäude- und Wohnungsidentifikatoren (EGID bzw. EWID) die wichtigsten Grunddaten wie z.B. Adresse, Standortkoordinaten, Baujahr, Anzahl Geschosse, Gebäudeart, Gebäudefläche, Heizungsart, Anzahl Wohnungen für die Gebäude geführt. Für die Wohnungen beispielsweise die Anzahl Zimmer und die Wohnungsfläche. Weitere Information zum GWR-Datenmodell finden Sie unter:

 - [Merkmalskatalog 4.2 (PDF)](https://www.housing-stat.ch/files/881-2200.pdf)
 - [Online Merkmalskatalog 4.2](https://www.housing-stat.ch/de/help/42.html)
 - oder generell auf der [GWR-Webseite des BFS](https://www.housing-stat.ch/de/index.html)

## Imports and helper functions

```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# helper function for reading datasets with proper separator
def get_dataset(url):
    """
    Return pandas df if url is parquet or csv file. Return None if not.
    """
    extension = url.rsplit(".",1)[-1]
    if extension == 'parquet':
        data = pd.read_parquet(url)
    elif extension == 'csv':
        data = pd.read_csv(url, sep=",", on_bad_lines='warn', encoding_errors='ignore', low_memory=False)
        # if dataframe only has one column or less the data is not comma separated, use ";" instead
        if data.shape[1] <= 1:
            data = pd.read_csv(url, sep=';', on_bad_lines='warn', encoding_errors='ignore', low_memory=False)
            if data.shape[1] <= 1:
                print("The data wasn't imported properly. Very likely the correct separator couldn't be found.\nPlease check the dataset manually and adjust the code.")
    else:
        print("Cannot load data! Please provide an url with csv or parquet extension.")
        data = None
    return data
```

```python
df = get_dataset('https://data.stadt-zuerich.ch/dataset/bau_gwr_wohnungen_best_od5503/download/BAU550OD5503.csv')
```
